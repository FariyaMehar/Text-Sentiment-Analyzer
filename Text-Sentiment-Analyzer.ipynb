```python
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return "Positive", polarity
    elif polarity < 0:
        return "Negative", polarity
    else:
        return "Neutral", polarity


def main():
    print("===== Text Sentiment Analyzer =====")

    while True:
        user_text = input("\nEnter text to analyze: ")

        sentiment, score = analyze_sentiment(user_text)

        print("\n--- Analysis Result ---")
        print("Input Text       :", user_text)
        print("Polarity Score   :", round(score, 2))
        print("Predicted Result :", sentiment)

        choice = input("\nAnalyze another text? (yes/no): ").lower()

        if choice != "yes":

            print("\nThank you for using Sentiment Analyzer!")
            break


if __name__ == "__main__":
    main()
```

```python
from textblob import TextBlob

def analyze_sentiment():
    print("--- AI Text Sentiment Analyzer ---")

    # 1. Accept user input text
    user_text = input("\nEnter the text you want to analyze: ")

    if not user_text.strip():
        print("Please enter some valid text.")
        return

    # 2. Use NLP techniques (via TextBlob)
    # TextBlob processes the string and calculates polarity
    blob = TextBlob(user_text)
    polarity = blob.sentiment.polarity

    # 3. Classify the sentiment
    # Polarity is a float within the range [-1.0, 1.0]
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    # 4. Display the predicted sentiment result
    print(f"\nAnalysis Result:")
    print(f"Text: '{user_text}'")
    print(f"Sentiment: {sentiment} (Score: {polarity:.2f})")

if __name__ == "__main__":
    analyze_sentiment()
```
